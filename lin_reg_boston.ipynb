## Creating Sample data for the model.

```python
X = raw_df.drop("PRICE", axis=1)
y = raw_df["PRICE"]
```

## Trian test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=43)
```

## Creating and training a Model

```python
from sklearn.linear_model import LinearRegression
```

```python
ob = LinearRegression()
```

```python
ob.fit(X_train, y_train)
```

```python
Y_pred = ob.predict(X_test)
```

## Model evaluation

```python
# print the intercept
print(ob.intercept_)
```

```python
# print the coefficient
print(ob.coef_)
```

```python
coeff_df = pd.DataFrame(ob.coef_,X.columns, columns = ["Coefficient"])
coeff_df 
```

```python
plt.scatter(y_test , Y_pred)
#plt.plot(X_test["CRIM"], Y_pred)
```

```python
from sklearn import metrics
print("Mean square error:", metrics.mean_squared_error(y_test,Y_pred) )
print("Mean absolute error:", metrics.mean_absolute_error(y_test,Y_pred))
print("Root mean square error(RMSE):",np.sqrt(metrics.mean_squared_error(y_test,Y_pred)))
```
